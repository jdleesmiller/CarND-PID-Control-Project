```python
%matplotlib inline
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('tmp') as f:
    data = f.read()
cte = []
for line in data.split("\n"):
    if len(line.split()) > 0:
        cte.append(float(line.split()[1]))
cte = np.array(cte)
```

```python
fs = 1 / 0.05
f, Pxx_den = signal.periodogram(cte, fs)
plt.semilogy(f, Pxx_den)
plt.xlim([0, 4])
plt.ylim([1e-7, 1e2])
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.show()
```
